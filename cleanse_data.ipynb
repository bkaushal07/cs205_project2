```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/Wholesale_customers_data.csv')
```

```python
data.head()
```

```python
cols = list(data.columns)
```

```python
data = data[cols]
data
```

```python
data.drop(['Region'], axis=1, inplace=True) #Nominal value
data.head()
```

```python
data.info()
```

```python
data.isnull().any()
```

```python
data.isna().sum()
```

```python
df_min_max_scaled = data.copy()

# apply normalization techniques by Column 1
column = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']
df_min_max_scaled[column] = (df_min_max_scaled[column] - df_min_max_scaled[column].min()) / (df_min_max_scaled[column].max() - df_min_max_scaled[column].min())

# view normalized data
display(df_min_max_scaled)
```

```python
df_min_max_scaled.to_csv('/Wholesale_customers_data_cleaned.csv', index=False, header=False, sep=' ')
```
